```python
x_cartao = base_cartao.iloc[:, [1,25]].values
x_cartao
```

```python
# Padronizar valores (K-means necessita)
from sklearn.preprocessing import StandardScaler
```

```python
scaler_cartao = StandardScaler()
x_cartao = scaler_cartao.fit_transform(x_cartao)
```

```python
x_cartao
```

```python
# definir o nr. de cluster visualizando o dendograma
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
dendrograma = dendrogram(linkage(x_cartao, method='ward'))
plt.title('Dendograma')
plt.xlabel('Pessoas')
plt.ylabel('Distância')
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc_cartao = AgglomerativeClustering(n_clusters=3, linkage='ward') # hc: hierarquical clustering

rotulos = hc_cartao.fit_predict(x_cartao)
```

```python
rotulos
```

```python
grafico1 = px.scatter(x=x_cartao[:,0], y=x_cartao[:,1], color=rotulos, width=1100, height=600)
grafico1.show()
```
